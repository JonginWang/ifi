```python
import os  # 1) 먼저 설정
# os.environ['NUMBA_THREADING_LAYER'] = 'tbb'
# (선택) OpenMP를 쓸 거라면
os.environ['NUMBA_THREADING_LAYER'] = 'omp'

# 2) 그 다음에 numba import
from numba import njit, prange, threading_layer
try:
    print("Layer:", threading_layer())
except ValueError:
    print("Layer: not initialized yet")

import numpy as np

@njit(parallel=True)
def f(a, b):
    for i in prange(a.size):
        a[i] += b[i]
    return a

a = np.ones(1_000_000); b = np.ones_like(a)
f(a, b)

print("After compile:", threading_layer())   # 'tbb' 나오면 정상
print(f"a[0]: {a[0]}")  # 2.0 나오면 정상
```
