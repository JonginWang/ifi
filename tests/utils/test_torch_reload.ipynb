# PyTorch reload after cache setup for this package

This notebook scripts to re-import the real PyTroch by the function `load_real_torch()`.

## Objectives:
1. **Reload of PyTorch**: Futher use of the torch after the fix of caches or `[WinErr 1114]`.
2. **Versatility of the funciton**: `load_real_torch()` can be used in the scipts, IDE, or REPL.

## Features:
- Cehck the existance of an attributes, `__version__`
- Remove the loaded PyTorch as an empty module.

```python
from ifi.utils.cache_setup import enable_torch
# Load torch
try:
    torch = enable_torch()
    print(f"torch.__version__: {torch.__version__}")
except Exception as e:
    print(f"âœ— Failed to load torch: {e}")
    print("  If torch is already loaded elsewhere, you may need to restart the kernel.")
    torch = None
```
